### Part 1: Chirp signal simulation/ pulse compression via matched filtering/basebanding etc.

In part 1 we assume that our sonar operates in air with a centre frequency of 10 kHz, with pulse bandwidth of 4 kHz. We will also use a sampling frequency of 44100 Hz. 

#### STEP 1: Chirp pulse creation

We first create a chirp pulse of the form 

$$v_{tx}(t) = rect(t/T)cos(2\pi[f_0+0.5Kt^2])$$ 


```julia
c = 343; # Speed of sound in air in m/s
fs = 44100; # This is the sample rate of the sonar.
dt = 1/fs; # This is the sample spacing
r_max = 4; # Maximum range in metres to which to simulate.
t_max = 2*r_max/c; # Time delay to max range

t = collect(0:dt:t_max); # t=0:dt:t_max defines a “range”.
r = c*t/2;

f0 = 10000; # Centre frequency is 10 kHz
B = 4000; # Chirp bandwidth
T = 6E-3; # Chirp pulse length
K = B/T; # Chirp rate

rect(t) = (abs.(t) .<= 0.5)*1.0
td = 0.6*T; # Chirp delay

v_tx = cos.( 2*pi*(f0*(t .- td) + 0.5*K*(t .- td).^2) ) .* rect.((t .-td)/T); 
```

```julia
using PyPlot # import plot library
figure(); # Create a new figure
title("Chirp waveform in time domain");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.1: Chirp Pulse ", fontsize=11, ha="center");
plot(t,v_tx); # Basic plot, axis labeled in samples

figure(); # Create a new figure
title("Chirp waveform against distance");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.2: Chirp Pulse ", fontsize=11, ha="center");
plot(r,v_tx); # Basic plot, axis labeled in samples
```

The following plots show a magnified version against time and distance.

```julia
figure();
title("Magnified Chirp Pulse")
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.3: Magnified Chirp Pulse", fontsize=11, ha="center")

plot(t[1:300],v_tx[1:300]); # Put time on x-axis
```

We can see that the sinusoid does not have a constant amplitude. This is because of the low sampling frequency causing loss of information. 

```julia
using FFTW 
V_TX = fft(v_tx); 
```

```julia
figure();
title("Frequency domain of chirp pulse")
figtext(.5,-0.05,"Figure 1.4: Frequency domain chirp pulse", fontsize=11, ha="center")
plot(fftshift(abs.(V_TX)));
```

```julia
figure();
title("Magnified Frequency domain chirp pulse")
figtext(.5,-0.05,"Figure 1.5: Magnified Frequency domain chirp pulse", fontsize=11, ha="center")
plot(abs.(V_TX)[100:350]);
```

We will now add the correct frequencies to the plot.

```julia
N = length(t);
Δf = 1/(N*dt) # spacing in frequency domain
f_axis = (-N/2:N/2-1)*Δf;

figure();
title("Frequency domain chirp pulse")
figtext(.5,-0.05,"Figure 1.6: Frequency domain chirp pulse", fontsize=11, ha="center")
plot(f_axis, abs.(fftshift(V_TX)),".-");
xlabel("Frequency in Hz");
```

Now we can see the frequency components around $10 kHz$ 

#### STEP 2: Point target simulation

We can simulate the received chirp pulse after it has bounced off an object at a distance R. To do this we simply multiply the transmitted signal by $1/R^2$ to simulated the attenuation, and move the waveform by $t_d = 2R/c$ to simulate the time it takes to propegate to the object and back. 

```julia
R1 = 1.5 +(12-9)/12 # My range to target.
td1 = 2*R1/c; # two way delay to target.
A1 = 1/R1^2; # echo voltage signal proportional to 1/R^2
v_rx = A1*cos.(2*pi*(f0*(t.-td.-td1)+0.5*K*(t.-td.-td1).^2)).*rect((t.-td.-td1)/T); 

figure();
title("Received chirp waveform")
figtext(.5,-0.05,"Figure 1.7: Received waveform", fontsize=11, ha="center")
xlabel("time (s)");
plot(t,v_rx);

figure();
title("Received chirp waveform against distance")
figtext(.5,-0.05,"Figure 1.8: Received waveform", fontsize=11, ha="center")
xlabel("distance (m)");
plot(r.*2,v_rx);
```

From figure 1.7 we can see the wave takes 0.1 seconds to return and from figure 1.8 we can see it travels around 3.5 m. 

```julia
V_RX = fft(v_rx);

N = length(t);
Δf = 1/(N*dt) # spacing in frequency domain
f_axis = (-N/2:N/2-1)*Δf;
figure();
title("Frequency domain of received waveform")
figtext(.5,-0.05,"Figure 1.9: Frequency domain of Received waveform", fontsize=11, ha="center")
plot(f_axis, abs.(fftshift(V_RX)),".-");
xlabel("Frequency in Hz");
```

We can simulate the received waveform after the transmitted waveform has bounced off two seperate objects by attenuating and shifting the transmitted pulse first with $R_1$ and then with $R_2$ and adding the waveforms together. 

```julia
R2 = 1.5 +(12-3)/12 # My range to target.
td2 = 2*R2/c; # two way delay to target.
A2 = 1/R2^2; # echo voltage signal proportional to 1/R^2

v_rx = A1*cos.(2*pi*(f0*(t.-td.-td1) + 0.5*K*(t.-td.-td1).^2)).*rect((t.-td.-td1)/T) + A2*cos.(2*pi*(f0*(t.-td.-td2) + 0.5*K*(t.-td.-td2).^2)).*rect((t.-td.-td2)/T);
V_RX = fft(v_rx);

figure();
title("Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.10: Received waveform with two objects", fontsize=11, ha="center")
xlabel("time (s)");
plot(t,v_rx); # Put range on x-axis

figure();
title("Magnigied Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.11: Received waveform with two objects", fontsize=11, ha="center")
xlabel("time (s)");
plot(t[450:900],v_rx[450:900]); # Put range on x-axis
```

We can see how there is constructive and destructive interference in the middle of the waveform where the two received waveforms overlap.

We can also see that the second waveform has a lower amplitude due to the longer distance it travels which causes it to attenuate further.

```julia
N = length(t);
Δf = 1/(N*dt) # spacing in frequency domain
f_axis = (-N/2:N/2-1)*Δf;
figure();
title("Frequency domain of Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.12: Frequency domain of Received waveform with two objects", fontsize=11, ha="center")
plot(f_axis, abs.(fftshift(V_RX)),".-");
xlabel("Frequency in Hz");

figure();
title("Magnified Frequency domain of Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.13:Frequency domain of Received chirp waveform with two objects", fontsize=11, ha="center")
plot(f_axis[100:400], abs.(fftshift(V_RX))[100:400],".-");
xlabel("Frequency in Hz (Zoomed in)");
```

#### STEP 3: Matched filtering

We an create a matched filter $$H(f) = conj(V_{ptr}(f))$$ matched to the response $v_{ptr}(t)$ that would be
recorded from a point target. 

We also assume that the imaginary component of the time domain output of the matched filter is zero

```julia
H = conj(V_TX);
```

```julia
V_MF = H.*V_RX;
v_mf = ifft(V_MF);
```

```julia

figure();
title("Matched filer response of Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.14: Matched filer response of Received chirp waveform with two objects", fontsize=11, ha="center")
xlabel("time (s)");
plot(t,v_mf);

figure();
title("Magnified Matched filer response of Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.15: Matched filer response of Received chirp waveform with two objects", fontsize=11, ha="center")
xlabel("time (s)");
plot(t[350:650],v_mf[350:650]);
```

We can clearly see the effect of pulse compression, the pulse widths are much smaller and there is much less overlap. 
We can also see that the second pulse has a lower amplitude, this is because the second pulse travels further and therefore attenuates more than the first. 

#### STEP 4: Forming an analytic signal

We will now find the analytic signal by removing the second half of the frequency domain array. 

```julia
V_ANAL = 2*V_MF; # make a copy and double the values
N = length(V_MF);

if mod(N,2)==0 # case N even
 neg_freq_range = Int(N/2):N; # Define range of “neg-freq” components
else # case N odd
 neg_freq_range = Int((N+1)/2):N;
end

V_ANAL[neg_freq_range] .= 0; # Zero out neg components in 2nd half of array.
v_anal = ifft(V_ANAL);
V_ANAL = 2*V_MF; # make a copy and double the values
N = length(V_MF);

if mod(N,2)==0 # case N even
 neg_freq_range = Int(N/2):N; # Define range of “neg-freq” components
else # case N odd
 neg_freq_range = Int((N+1)/2):N;
end

V_ANAL[neg_freq_range] .= 0; # Zero out neg components in 2nd half of array.
v_anal = ifft(V_ANAL);
```

```julia
figure()
title("Analytic version of received waveform");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.16: Analytic version of received waveform", fontsize=11, ha="center");
plot(r,abs.(v_anal));

figure();
title("Magnified analytic version of received waveform");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.17: magnified analytic version of received waveform", fontsize=11, ha="center");
plot(r[445:457],abs.(v_anal)[445:457]);

figure();
title("real part of analytic version of received waveform");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.18: real part of analytic version of received waveform", fontsize=11, ha="center");
plot(r,real(v_anal));
```

If we magnify the analytic signal so that we can only see the main lobe of the first impulse, we can see that the $3 dB$ point lies between the points where $y = 42.5*0.707 = 30$. Therefore the $3 db$ width is around $0.04$ which agrees with our theory since $\frac{343}{2*4000} = 0.043$

We can also see that if we remove the imaginary component of the analytic signal we get the original signal back. 

#### STEP 5: Translating the signal to baseband

We can translate the signal to baseband by multiplying the time-domain analytic signal by $e^{-j2\pi f_0t}$

```julia
j=im; # Assign j as sqrt(-1) (“im” in julia)
v_bb = v_anal.*exp.(-j*2*pi*f0*t);
```

```julia
figure();
subplot(2,1,1);
plot(r,abs.(v_bb));
figure();
subplot(2,1,2);
plot(r,angle.(v_bb));

figure();
subplot(2,1,1);
plot(r[400:500],abs.(v_bb)[400:500]);
figure();
subplot(2,1,2);
plot(r[400:500],angle.(v_bb)[400:500]);
```

We can see that the phase is constant over the main lobe, and side lobes to a lesser extent. 

```julia
V_BB = fft(v_bb)

figure();
title("Frequency domain of baseband signal");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.19: Frequency domain of baseband signal", fontsize=11, ha="center");
plot(abs.(V_BB));

figure();
title("Magnified frequency domain of baseband signal");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.20: Magnified frequency domain of baseband signal", fontsize=11, ha="center");
plot(abs.(V_BB[1:100]));
```

When plotting the fourier transform of the analytic signal the centre frequeny lies at around 0 Hz and the bandwidth is split between the start and end of the array. This is due to the repeating nature of the basebanded signal, the end of the array is the start of a new "repeat"

#### Step 7: Compensating for range dependence

Two identical targets placed at different distances will appear as peaks of different heights. To compensate for the $\frac{1}{R^2}$ attenuation we can multiply the final array with the distance array squared, therefore adding a quadratic gain to the array. 

```julia
R1 = 1.5 +(12-9)/12 # My range to target.
td1 = 2*R1/c; # two way delay to target.
A1 = 1/R1^2; # echo voltage signal proportional to 1/R^2

R2 = 1.5 +(12-3)/12 # My range to target.
td2 = 2*R2/c; # two way delay to target.
A2 = 1/R2^2; # echo voltage signal proportional to 1/R^2

v_rx = A1*cos.(2*pi*(f0*(t.-td.-td1) + 0.5*K*(t.-td.-td1).^2)).*rect((t.-td.-td1)/T) + A2*cos.(2*pi*(f0*(t.-td.-td2) + 0.5*K*(t.-td.-td2).^2)).*rect((t.-td.-td2)/T);
V_RX = fft(v_rx);
H = conj(V_TX);
V_MF = H.*V_RX;
v_mf = ifft(V_MF);
v_mf = real(v_mf);
r_squared = r.*r;

figure();
title("Time domain plot of compensated signal");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.23: Time domain plot of compensated signal", fontsize=11, ha="center");
plot(t,r_squared.*v_mf);
```

We can see now that the two pulses are at around the same height.

#### Step 9 – Sampling below the usual Nyquist rate

The Nyquist theorem for bandpass signals states that one can retain all information in a sampled
bandlimited signal by sampling at k*B where 2<k<=4. We will change $f_s$ to $4B$ to see how this affects the signal.

```julia
c = 343; # Speed of sound in air in m/s
fs = 100000; # This is the sample rate of the sonar.
dt = 1/fs; # This is the sample spacing
r_max = 4; # Maximum range in metres to which to simulate.
t_max = 2*r_max/c; # Time delay to max range

t = collect(0:dt:t_max); # t=0:dt:t_max defines a “range”.
r = c*t/2;

f0 = 10000; # Centre frequency is 10 kHz
B = 4000; # Chirp bandwidth
T = 6E-3; # Chirp pulse length
K = B/T; # Chirp rate

td = 0.6*T; # Chirp delay
v_tx = cos.( 2*pi*(f0*(t .- td) + 0.5*K*(t .- td).^2) ) .* rect.((t .-td)/T); 

figure()
title("Magnified Time domain plot with high sampling rate");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.30: Magnified Time domain plot with high sampling rate", fontsize=11, ha="center");
plot(t[1:700], v_tx[1:700]);

c = 343; # Speed of sound in air in m/s
fs = 4*4000; # This is the sample rate of the sonar.
dt = 1/fs; # This is the sample spacing
r_max = 4; # Maximum range in metres to which to simulate.
t_max = 2*r_max/c; # Time delay to max range

t = collect(0:dt:t_max); # t=0:dt:t_max defines a “range”.
r = c*t/2;

f0 = 10000; # Centre frequency is 10 kHz
B = 4000; # Chirp bandwidth
T = 6E-3; # Chirp pulse length
K = B/T; # Chirp rate

td = 0.6*T; # Chirp delay
v_tx = cos.( 2*pi*(f0*(t .- td) + 0.5*K*(t .- td).^2) ) .* rect.((t .-td)/T); 

figure();
title("Magnified Time domain plot with low sampling rate");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.31: Magnified Time domain plot with low sampling rate", fontsize=11, ha="center");
plot(t[1:120], v_tx[1:120]);

```

If we use $4*4000$ as the sampling frequency we clearly lose a lot of the information in the signal, however, we should be using $4*f_c = 4*(10000+B/2)$, this gives us a closer representation of the signal as shown below.

```julia
c = 343; # Speed of sound in air in m/s
fs = 4*(10000+2000); # This is the sample rate of the sonar.
dt = 1/fs; # This is the sample spacing
r_max = 4; # Maximum range in metres to which to simulate.
t_max = 2*r_max/c; # Time delay to max range

t = collect(0:dt:t_max); # t=0:dt:t_max defines a “range”.
r = c*t/2;

f0 = 10000; # Centre frequency is 10 kHz
B = 4000; # Chirp bandwidth
T = 6E-3; # Chirp pulse length
K = B/T; # Chirp rate

td = 0.6*T; # Chirp delay
v_tx = cos.( 2*pi*(f0*(t .- td) + 0.5*K*(t .- td).^2) ) .* rect.((t .-td)/T); 

figure();
title("Magnified Time domain plot with 4*(fc+B/2) sampling rate");
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.32: Magnified Time domain plot with 4*(fc+B/2) sampling rate", fontsize=11, ha="center");
plot(t[1:350], v_tx[1:350]);
```
