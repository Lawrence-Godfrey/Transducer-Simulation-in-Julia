### Part 1: Chirp signal simulation/ pulse compression via matched filtering/basebanding etc.

In part 1 we assume that our sonar operates in air with a centre frequency of 10 kHz, with pulse bandwidth of 4 kHz. We will also use a sampling frequency of 44100 Hz. 

#### STEP 1: Chirp pulse creation

We first create a chirp pulse of the form 

$$v_{tx}(t) = rect(t/T)cos(2\pi[f_O+0.5Kt^2])$$ 


```julia
c = 343; # Speed of sound in air in m/s
fs = 44100; # This is the sample rate of the sonar.
dt = 1/fs; # This is the sample spacing
r_max = 4; # Maximum range in metres to which to simulate.
t_max = 2*r_max/c; # Time delay to max range

# Create an array containing the time values of the samples
t = collect(0:dt:t_max); # t=0:dt:t_max defines a “range”.
# Create an array containing the range values of the samples
r = c*t/2;

# NOW create the chirp pulse, shifted by an amount td, to start at
# some time td-T/2>0.
f0 = 10000; # Centre frequency is 10 kHz
B = 4000; # Chirp bandwidth
T = 6E-3; # Chirp pulse length
K = B/T; # Chirp rate

# Define a simple a rect() function which returns for -0.5<=t<=0.5 or 0.
# The function will work if t is an array of values.
rect(t) = (abs.(t) .<= 0.5)*1.0

# rect(t/T) spans the interval [-T/2,T/2]
# We must therefore delay the chirp pulse so that it starts after t=0.
# Shift the chirp pulse by 0.6T units to the right, so that it starts at 0.1*T
td = 0.6*T; # Chirp delay

# Note: one can use the macro “@.” to avoid having to put “.” for arrays:
# @. v_tx = cos( 2*pi*(f0*(t-td) + 0.5*K*(t-td).^2) ).*rect((t-td)/T);
v_tx = cos.( 2*pi*(f0*(t .- td) + 0.5*K*(t .- td).^2) ) .* rect.((t .-td)/T); 
```

Examine the signal in time and frequency domains, using the plot commands:

```julia
using PyPlot # import plot library
figure() # Create a new figure
title("Chirp waveform in time domain")
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.1: Chirp Pulse ", fontsize=11, ha="center")
plot(v_tx[1:300]) # Basic plot, axis labeled in samples


f = figure()
     withfig(f) do #removes output text
end;
```

```julia
figure()
title("Magnified Chirp Pulse")
xlabel("time (s)");
figtext(.5,-0.05,"Figure 1.2: Magnified Chirp Pulse", fontsize=11, ha="center")

plot(t[1:300],v_tx[1:300]) # Put time on x-axis

figure()
title("Magnified Chirp pulse")
xlabel("distance (m)");
figtext(.5,-0.05,"Figure 1.3: Magnified Chirp pulse", fontsize=11, ha="center")

plot(r[1:300],v_tx[1:300]) # Put range on x-axis

f = figure()
     withfig(f) do #removes output text
end;
```

We can see that the sinusoid does not have a constant amplitude. This is because of the low sampling frequency causing loss of information. 

```julia
using FFTW 
V_TX = fft(v_tx); 
```

```julia
figure();
title("Frequency domain chirp pulse")
figtext(.5,-0.05,"Figure 1.4: Frequency domain chirp pulse", fontsize=11, ha="center")
plot(abs.(V_TX));
```

```julia
figure();
title("Magnified Frequency domain chirp pulse")
figtext(.5,-0.05,"Figure 1.5: Magnified Frequency domain chirp pulse", fontsize=11, ha="center")
plot(abs.(V_TX)[100:400]);
```

We will now add the correct frequencies to the plot.

```julia
#LABEL frequency axis
N = length(t);
Δf = 1/(N*dt) # spacing in frequency domain
#create array of freq values stored in f_axis. First element maps to 0Hz
f_axis = (0:N-1)*Δf;
figure();
title("Frequency domain chirp pulse")
figtext(.5,-0.05,"Figure 1.6: Frequency domain chirp pulse", fontsize=11, ha="center")
plot(f_axis, abs.(V_TX),".-");
xlabel("Frequency in Hz");
```

#### STEP 2: Point target simulation

We can simulate the received chirp pulse after it has bounced off an object at a distance R. To do this we simply multiply the transmitted signal by $1/R^2$ to simulated the attenuation, and move the waveform by $t_d = 2R/c$ to simulate the time it take to propegate to the object and back. 

```julia
R1 = 1.5 +(12-9)/12 # My range to target.
td1 = 2*R1/c; # two way delay to target.
A1 = 1/R1^2; # echo voltage signal proportional to 1/R^2
v_rx = A1*cos.( 2*pi*(f0*(t.-td.-td1) + 0.5*K*(t.-td.-td1).^2) ) .* rect((t.-td.-td1)/T); 

figure()
title("Received chirp waveform")
figtext(.5,-0.05,"Figure 1.7: Received waveform", fontsize=11, ha="center")
xlabel("time (s)");
plot(t,v_rx) # Put range on x-axis

f = figure()
     withfig(f) do #removes output text
end;
```

```julia
figure()
title("Magnified Received chirp waveform")
figtext(.5,-0.05,"Figure 1.8: Magnified Received waveform", fontsize=11, ha="center")
xlabel("time (s)");
plot(t[300:800],v_rx[300:800]) # Put range on x-axis

f = figure()
     withfig(f) do #removes output text
end;
```

```julia
V_RX = fft(v_rx);

#LABEL frequency axis
N = length(t);
Δf = 1/(N*dt) # spacing in frequency domain
#create array of freq values stored in f_axis. First element maps to 0Hz
f_axis = (0:N-1)*Δf;
figure();
title("Frequency domain of received waveform")
figtext(.5,-0.05,"Figure 1.9: Frequency domain of Received waveform", fontsize=11, ha="center")
plot(f_axis, abs.(V_RX),".-");
xlabel("Frequency in Hz");
```

We can simulate the received waveform after bouncing off two seperate objects by attenuating and shifting the transmitted pulse first with &R_1& and then with &R_2& and adding the waveforms together. 

```julia
R2 = 1.5 +(12-3)/12 # My range to target.
td2 = 2*R2/c; # two way delay to target.
A2 = 1/R2^2; # echo voltage signal proportional to 1/R^2

v_rx = A1*cos.(2*pi*(f0*(t.-td.-td1) + 0.5*K*(t.-td.-td1).^2)).*rect((t.-td.-td1)/T) + A2*cos.(2*pi*(f0*(t.-td.-td2) + 0.5*K*(t.-td.-td2).^2)).*rect((t.-td.-td2)/T);
V_RX = fft(v_rx);

figure();
title("Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.10: Received waveform with two objects", fontsize=11, ha="center")
xlabel("time (s)");
plot(t,v_rx); # Put range on x-axis

figure();
title("Magnigied Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.11: Received waveform with two objects", fontsize=11, ha="center")
xlabel("time (s)");
plot(t[450:900],v_rx[450:900]); # Put range on x-axis
```

We can see how there is constructive and destructive interference in the middle of the waveform where the two received waveforms overlap.

```julia
#LABEL frequency axis
N = length(t);
Δf = 1/(N*dt) # spacing in frequency domain
#create array of freq values stored in f_axis. First element maps to 0Hz
f_axis = (0:N-1)*Δf;
figure();
title("Frequency domain of Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.12: Frequency domain of Received waveform with two objects", fontsize=11, ha="center")
plot(f_axis, abs.(V_RX),".-");
xlabel("Frequency in Hz");

figure();
title("Magnified Frequency domain of Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.13:Frequency domain of Received chirp waveform with two objects", fontsize=11, ha="center")
plot(f_axis[100:400], abs.(V_RX)[100:400],".-");
xlabel("Frequency in Hz (Zoomed in)");
```

#### STEP 3: Matched filtering

We an create a matched filter $$H(f) = conj(V_{ptr}(f))$$ matched to the response $v_{ptr}(t)$ that would be
recorded from a point target. 

```julia
H = conj(V_TX);
```

Apply the matched filter to the simulated return and plot the output in time and frequency domains.

```julia
V_MF = H.*V_RX;
v_mf = ifft(V_MF);
```

NOTE: If all has worked correctly, then in v_mf should ideally be a purely real signal. In some
implementations of fft and ifft, the time domain signal may have a tiny imaginary component.
 introduced by numerical error in the fft and ifft processing. You can check this by plotting: 

```julia

figure();
title("Matched filer response of Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.14: Matched filer response of Received chirp waveform with two objects", fontsize=11, ha="center")
xlabel("time (s)");
plot(real(v_mf));

figure();
title("Magnified Matched filer response of Received chirp waveform with two objects")
figtext(.5,-0.05,"Figure 1.15: Matched filer response of Received chirp waveform with two objects", fontsize=11, ha="center")
xlabel("time (s)");
plot(real(v_mf[350:650]));
```

We an see that the second pulse has a lower amplitude, this is due to pulse compression. The second pulse travels further and therefore attenuates more than the first. 

#### STEP 4: Forming an analytic signal

We will now find the analytic signal by removing the second half of the frequency domain array. 

```julia
V_ANAL = 2*V_MF; # make a copy and double the values
N = length(V_MF);

if mod(N,2)==0 # case N even
 neg_freq_range = Int(N/2):N; # Define range of “neg-freq” components
else # case N odd
 neg_freq_range = Int((N+1)/2):N;
end

V_ANAL[neg_freq_range] .= 0; # Zero out neg components in 2nd half of array.
v_anal = ifft(V_ANAL);
```

#### Step 8: Experimenting with the sample rate

Sonar signals are usually represented as baseband signals in I/Q format. The final processed output
is the envelope of the pulse-compressed (via inverse filter or matched filter) range profile. The
sample rate is not usually high enough to produce a smooth looking point target response. Thus it is
desirable to increase the sample rate in the time domain.

You could try to increase the ADC sample rate fs, to allow one to see smoother waveforms, if you
zoom in to the 'RF' time domain waveforms.
An alternative method is to implement the technique known as “zero-padding” in the frequency
domain, in order to produce more closely spaced samples in the time domain. The IFFT maps N
samples into N samples. The idea is to increase the size of the frequency domain array by a factor
(of say 10) by inserting additional 9N zeros into the centre of the array.
[12345678] --→ [ 1234 00000….0000 5678 ]
Method: Create a new array containing all zeros ( e.g. zeros(10*N) ) and copy the first half of the
Fourier spectrum into the first 1:N/2 samples and the 2nd half of the array into the last part of the
new array (into position ~10N-N/2+1:10*N). Then take the inverse transform. The resulting time
domain will have a finer spacing. Plotted waveforms will look smoother. The time/range axis will
have to be relabelled with the new finer sample spacing.

```julia
V_TX_new = V_TX

figure()
plot(abs.(V_TX_new)[100:350],".-");
xlabel("Frequency in Hz");

figure()
v_tx_new = ifft(V_TX_new);
plot(v_tx_new[1:200])

padding = zeros(10N)

V_TX_new = abs.(V_TX_new)

padding[1:round(Int,(N/2))]=V_TX_new[1:round(Int,(N/2))]
padding[10N - round(Int,(N/2)+1):10N]=V_TX_new[round(Int,(N/2))-1:N]

figure()
plot(abs.(padding[1:500]),".-");
xlabel("Frequency in Hz");

figure()
v_tx_new_padded = ifft(padding);
plot(real(v_tx_new_padded)[1:2000])


```
